```python
import pandas as pd
import plotly.express as px
```

```python
df_incoming_run = pd.read_parquet("train/incoming_run_data_1.parquet")

df_incoming_run.head()
```

```python
df_incoming_run.info()
```

```python
df_incoming_run.describe()
```

```python
for sensor_name in df_incoming_run["Sensor Name"].unique():
    fig = px.histogram(
        df_incoming_run[df_incoming_run["Sensor Name"] == sensor_name],
        x="Sensor Value",
        title=sensor_name,
    )
    fig.show()
```
